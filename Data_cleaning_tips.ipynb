```python
import pandas as pd
import numpy as np
```

```python
url='https://github.com/fivethirtyeight/data/blob/master/comic-characters/marvel-wikia-data.csv?raw=true'# data form https://github.com/andrewflowers 
comic=pd.read_csv(url)
comic
```

The three common operation I do while cleaning data.

```python
def data_format(s):
  s=s.lower()# set all character to lower case
  s=s.strip()#If you want to remove leading and ending spaces
  s=s.replace("'",'')#replace dobule quotation 
  s=s.replace('"','')#replace quotation 
  return(s)
```

How to use data format function on pandas set use apply and lambda

```python
comic['output']=comic['name'].apply(lambda x: data_format(x))
comic
```

Changing the case in a dataframe
