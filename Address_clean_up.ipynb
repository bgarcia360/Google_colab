Function that clean up the address using regex with boundary.

```python
def add_clean(address):
  import re
  address = re.sub("[^a-zA-Z0-9 ]+","",address) 
  address =address.lower() 
  
  toReplace = {"st": 'Street', "rd": "Road", "ave": "Avenue","dr":"Drive","blvd":"BOULEVARD", "w":"West", "n":"North","e":"East","s":"South"} 
  for k,v in toReplace.items():
    address = re.sub(r"\b" + k + r"\b", v, address)

  address=address.strip().title()
  
  return address
```

Using the Lambda and apply to apply to append to data frame

```python
address['ouput']= address['address'].apply(lambda x: add_clean(x))

address
```
